```python
# Q2 (3)
titanic_df.isna().sum()
```

```python
# Q2 (4)
titanic_df_drop = titanic_df.drop(columns=['Cabin'])
titanic_df_drop
```

```python
# Q2 (5)
titanic_df_drop.loc[titanic_df_drop.Age.isna()]
```

```python
# Q2 (6)
titanic_df_drop['Age'] = titanic_df_drop['Age'].fillna(titanic_df_drop['Age'].mean())
titanic_df_drop
```

```python
# Q2 (7)
titanic_df_drop['Embarked'] = titanic_df_drop['Embarked'].fillna(titanic_df_drop['Embarked'].mode())
titanic_df_drop
```

```python
# Q2 (8)
print('male:',titanic_df_drop[titanic_df_drop['Sex']=='male']['Age'].mean())
print('female:',titanic_df_drop[titanic_df_drop['Sex']=='female']['Age'].mean())
```

```python
# Q2 (9)
titanic_df_drop.loc[(titanic_df_drop['Sex'] == 'male') & (titanic_df_drop['Age'] < 12), 'Fare'] *= 0.8
titanic_df_drop['Fare']
```

```python
# Q2 (10)
titanic_df_drop.loc[titanic_df_drop['Age'] > 40, 'Fare'] = titanic_df_drop.loc[titanic_df_drop['Age'] > 40, 'Fare'] * 0.9
titanic_df_drop.loc[titanic_df_drop['Age'] > 40, 'Fare'] = titanic_df_drop.loc[titanic_df_drop['Age'] > 40, 'Fare'].round(1)
titanic_df_drop['Fare']
```

```python
# Q3 (8)
cylinder_mapping = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}
automobile_data['num-of-cylinders'] = automobile_data['num-of-cylinders'].replace(cylinder_mapping)
automobile_data
```

```python
# Q3 (9)
automobile_data = automobile_data.assign(ratio_of_price_cylinders = automobile_data['price'] / automobile_data['num-of-cylinders'])
automobile_data
```

```python
# Q3 (10)
df_audi = automobile_data.loc[(automobile_data['company'] == 'audi') & (automobile_data['body-style'] == 'sedan')]
df_bmw = automobile_data.loc[(automobile_data['company'] == 'bmw') & (automobile_data['body-style'] == 'sedan')]
merged_df = pd.concat([df_audi, df_bmw], axis=0)
merged_df
```
