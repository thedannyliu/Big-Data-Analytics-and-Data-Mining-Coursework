```python
# Q2-1

data = {'Book ID': ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B9', 'B10'],
        'Income': [1000, 3000, 30000, 24000, 300, 400, 800, 2300, 12000, 1800]}

df = pd.DataFrame(data)
df
```

```python
# Q2-2

df_sorted = df.sort_values(by='Income', ascending=False).reset_index(drop=True)
df_sorted
```

```python
# Q2-3

total_income = df_sorted['Income'].sum()
df_sorted['Income_percent'] = df_sorted['Income'] / total_income
df_sorted
```

```python
# Q2-4

df_sorted['Income_cum_percent'] = df_sorted['Income_percent'].cumsum()
df_sorted
```

```python
# Q2-5

# create figure and axes objects
fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.set_facecolor('lightgray')
ax2 = ax1.twinx()

# create bar chart for income
ax1.bar(df_sorted['Book ID'], df_sorted['Income'], color='blue', zorder=2)

# create line chart for income cumulative percent
ax2.plot(df_sorted['Book ID'], df_sorted['Income_cum_percent'], color='red',linewidth=2)

# set labels and titles
ax1.set_xlabel('圖書ID', fontsize=14, fontname='SimHei')
ax1.set_ylabel('收入', fontsize=14, fontname='SimHei')
ax2.set_ylabel('收入累積比例', fontsize=14, fontname='SimHei')
ax1.grid(axis='y',color='white')
ax1.grid(axis='x',color='white')
ax2.grid(False)
ax1.tick_params(axis='y')
ax2.tick_params(axis='y')
plt.title('收入主次因素分析', fontsize=16, fontname='SimHei')

# display plot
plt.show()
```

```python
# Q3-1

df = pd.read_excel('C:/Users/danny/Big_Data_Analytics_and_Data_Mining/Homework 2/health.xlsx')
df
```

```python
# Q3-2

df['每單位卡路里之行走步數'] = df['每日行走步數'] / df['攝取卡路里']
df
```

```python
# Q3-3

def health_index(x):
    if x <= 2:
        return '低'
    elif 2 <= x <= 4.5:
        return '中'
    else:
        return '高'

df['健康指數'] = df['每單位卡路里之行走步數'].apply(health_index)
df
```

```python
# Q3-4

# One-hot Encoding
onehot = pd.get_dummies(df['健康指數'])

# Rename columns
onehot.columns = ['健康指數_低', '健康指數_中', '健康指數_高']

# Sort columns
onehot = onehot[['健康指數_高', '健康指數_中', '健康指數_低']]

# Concatenate with original DataFrame
df = pd.concat([df, onehot], axis=1)
df
```
